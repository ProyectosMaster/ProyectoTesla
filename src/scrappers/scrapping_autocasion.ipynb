```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
c_options = webdriver.ChromeOptions()
c_options.add_argument("--start-maximized")
c_driver = webdriver.Chrome(options=c_options)
time.sleep(3)
```

```python
# ---------------------------------------------------- COCHES CON COLOR -----------------------------------------------------------------
# from selenium import webdriver
# from selenium.webdriver.common.by import By
# import time

# # Iniciar el driver de Selenium
# c_options = webdriver.ChromeOptions()
# c_options.add_argument("--start-maximized")
# c_driver = webdriver.Chrome(options=c_options)


results = {
    "modelo": [],
    "precio": [],
    "kilometraje": [],
    "matriculacion": [],
    "motor": [],
    "color": []
}

# Accedemos a la URL
lista = [("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-azul?gad_source=1&utm_term=","azul"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-blanco?gad_source=1&utm_term=","blanco"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-granate?gad_source=1&utm_term=","granate"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-gris?gad_source=1&utm_term=","gris"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-marron?gad_source=1&utm_term=","marron"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-negro?gad_source=1&utm_term=","negro"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-plata?gad_source=1&utm_term=","plata"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-rojo?gad_source=1&utm_term=","rojo"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-azul/metalizado?gad_source=1&utm_term=","azul"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-blanco/metalizado?gad_source=1&utm_term=","blanco"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-granate/metalizado?gad_source=1&utm_term=","granate"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-gris/metalizado?gad_source=1&utm_term=","gris"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-marron/metalizado?gad_source=1&utm_term=","marron"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-negro/metalizado?gad_source=1&utm_term=","negro"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-plata/metalizado?gad_source=1&utm_term=","plata"),
         ("https://www.autocasion.com/coches-segunda-mano/tesla-ocasion/color-rojo/metalizado?gad_source=1&utm_term=","rojo")]
# url = "https://www.autocasion.com/coches-segunda-mano/tesla-ocasion?gad_source=1&utm_term="
  

for url, color in lista:
    c_driver.get(url)
    time.sleep(5)
# Navegación por cada anuncio
    while True:
        try:
            contenido = c_driver.find_elements(By.CLASS_NAME, "anuncio")
            if not contenido:
                print("No se encontraron anuncios en esta página.")
                break  # Salimos si no encontramos anuncios

            for elemento in contenido:
                try:
                    titulo = elemento.find_element(By.XPATH, "//h2[@itemprop='name']").text
                    modelo = titulo.split("TESLA", 1)[-1].strip()
                except Exception as e:
                    print(f"Error al extraer el modelo: {e}")
                    modelo = "Sin dato"

                try:
                    precio = elemento.find_element(By.CLASS_NAME, "precio").text
                    precio_final = precio.split("\n")[1]  # Esto nos da "21.990 € -6%" o "21.990 €"
                    precio_final = precio_final[:8]
                except Exception as e:
                    print(f"Error al extraer el precio: {e}")
                    precio_final = "Sin dato"

                try:
                    lista_items = elemento.find_elements(By.TAG_NAME, "ul")
                    for ul in lista_items:
                        li_textos = [li.text for li in ul.find_elements(By.TAG_NAME, "li")]
                        matriculacion = li_textos[0] if len(li_textos) > 0 else "Sin dato"
                        motor = li_textos[1] if len(li_textos) > 1 else "Sin dato"
                        kilometraje_texto = li_textos[2] if len(li_textos) > 2 else "Sin dato"
                        kilometraje = kilometraje_texto.replace("km", "").strip()
                except Exception as e:
                    print(f"Error al extraer la información de matrícula, motor o kilometraje: {e}")
                    matriculacion = "Sin dato"
                    motor = "Sin dato"
                    kilometraje = "Sin dato"

                # Almacenar los resultados
                results['modelo'].append(modelo)
                results['precio'].append(precio_final)
                results['kilometraje'].append(kilometraje)
                results['matriculacion'].append(matriculacion)
                results['motor'].append(motor)
                results['color'].append(color)

            # Intentamos hacer clic en el siguiente botón para la siguiente página
            try:
                siguiente = c_driver.find_element(By.CLASS_NAME, "last")
                time.sleep(5)
                siguiente.click()
            except Exception as e:
                print(f"Error al intentar hacer clic en el siguiente botón: {e}")
                break  # Salimos del bucle si no se encuentra el botón de siguiente

        except Exception as e:
            print(f"Error: {e}")
            break  
```

```python
df_prueba = pd.DataFrame(results)
df_prueba
```
